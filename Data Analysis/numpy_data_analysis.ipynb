```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

```python
url = "https://raw.githubusercontent.com/mwaskom/seaborn-data/master/titanic.csv"
df= pd.read_csv(url)
df.head()
```

```python
df.isnull().sum()
```

```python
df.dtypes
```

```python
age = df['age'].to_numpy()
fare = df['fare'].to_numpy()
```

```python
print("Average age:", np.nanmean(age))
print("Average fare:", np.nanmean(fare))

print("Youngest passenger age:", np.nanmin(age))
print("Oldest passenger age:", np.nanmax(age))
```

```python
print("Average Fare:", np.nanmean(fare))

print("Median Fare:", np.nanmedian(fare))

print("Lowest Fare:", np.nanmin(fare))
print("Highest Fare:", np.nanmax(fare))

print("Fare standard deviation:", np.nanstd(fare))
```

```python
survival_counts = df['survived'].value_counts()
print(survival_counts)
```

```python
survival_labels = ['Did Not Survive', 'Survived']
plt.figure(figsize=(6, 5))
plt.bar(survival_counts.index, survival_counts, color=['red', 'green'])
plt.ylabel('Number of Passengers')
plt.title('Survival Counts on the Titanic')
plt.tight_layout()
plt.show()
```

```python
survival_rate = np.mean(df['survived'].to_numpy()) * 100
print(f"Overall Survival Rate: {survival_rate:.2f}%")
```

```python
gender_survived = df.groupby('sex')['survived'].mean() * 100
print(gender_survived)
```

```python
plt.figure(figsize=(6,5))
plt.bar(gender_survived.index, gender_survived.values, color=['pink', 'blue'])
plt.title('Survival Rate by Gender')
plt.ylabel('Survival Rate (%)')
plt.ylim(0, 100)
plt.xlabel('Genders')
plt.tight_layout()
plt.show()
```
